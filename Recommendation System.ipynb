# Using Weighted avg Formula 

W = Weighted Rating

R  = Voting average of the movie

v = Number of total votes for the movie

m = Minimum votes required to be listed in top 250

C = The mean vote across the whole report


```python
R= movies_credits_merge['vote_average']
v= movies_credits_merge['vote_count']
C= movies_credits_merge['vote_average'].mean()
m= movies_credits_merge['vote_count'].quantile(.70)
```

```python
movies_credits_merge['Weighted_avg']= ((R*v)+(C*m))/(v+m)
```

```python
movies_credits_merge.head()
```

```python
movies_ranking = movies_credits_merge.sort_values('Weighted_avg',ascending=False)
```

```python
movies_ranking[['original_title','vote_count','vote_average','Weighted_avg','popularity']].head(20)
```

# Best Movies based on calculated weighted Avg

```python
weight_average=movies_ranking.sort_values('Weighted_avg',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['Weighted_avg'].head(10),y=weight_average['original_title'].head(10))
plt.xlim(4,10)
plt.title('Best Movies by Average votes',weight='bold')
plt.xlabel('weighted avg score', weight='bold')
plt.ylabel('movie title',weight='bold')
plt.savefig('best_movies.png')
```

# Best Movies based on Popularity

```python
popularity=movies_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(15,9))
axis1=sns.barplot(x=popularity['popularity'].head(10),y=popularity['original_title'].head(10),data=popularity)

plt.title('Best Movies based on popularity',weight='bold')
plt.xlabel('popularity', weight='bold')
plt.ylabel('movie title',weight='bold')
plt.savefig('best_movies_pupularity.png')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
movies_credits_merge['overview']=movies_credits_merge['overview'].fillna('')
```

```python
tfidf=TfidfVectorizer(strip_accents='ascii',stop_words='english',analyzer='word',token_pattern=r'\w{1,}',
                      ngram_range=(1,3),min_df=3)
```

```python
tfidf_matrix=tfidf.fit_transform(movies_credits_merge['overview'])
```

```python
tfidf_matrix
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
```

```python
sig = sigmoid_kernel(tfidf_matrix,tfidf_matrix)
```

```python
sig
```

```python
sig[0]
```

```python
indices=pd.Series(movies_credits_merge.index,index=movies_credits_merge['original_title']).drop_duplicates()
indices.head(15)
```

```python
indices['Jab Tak Hai Jaan']
```

```python
list(enumerate(sig[indices['Jab Tak Hai Jaan']]))
```

```python
sorted(list(enumerate(sig[indices['Jab Tak Hai Jaan']])), key=lambda x: x[1], reverse=True)
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return movies_credits_merge['original_title'].iloc[movie_indices]
```

```python
give_rec('Jab Tak Hai Jaan')
```
